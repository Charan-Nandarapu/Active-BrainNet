```python
#importing the necessary libraries
import os
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import random
import pickle
```

```python
#mounting drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
#defining the training path
train_path = '/content/drive/MyDrive/Research Study/trial Dataset/Dataset'
```

```python
#checking numbero available folders and printing their names
File=[]
for f in os.listdir(train_path):
    File += [f]

'''  total number of classes '''
print(File)
```

```python
train_data = []                                                                 #train data array
mapping = {'1_notumor': 0, '2_pituitary': 1, '3_meningioma': 2, '4_glioma': 3}  #defining the classes

count=0                                                                         #count variable for appeneind labels to the data

for f in os.listdir(train_path):                                                #looping over directories
    path = os.path.join(train_path, f)                                          #joining path
    for im in os.listdir(path):                                                 #looping through images in the folder
        img = load_img(os.path.join(path, im), grayscale=False, color_mode='rgb', target_size=(150,150)) #loading the image into a varibale
        img = img_to_array(img)                                                 #converting the image to numpy array
        img = img / 255.0                                                       #normalising the arrays to the range of 0 to 1
        train_data.append([img, count])                                         #appending the image array and label to one single list
    count=count+1                                                               #label value changes as it is going to next folder
```

```python
len(train_data) #length of the training data
```

```python
# Shuffle the data
random.shuffle(train_data)
```

```python
# Calculate the number of images to sample (3% of total)
num_images_to_sample = int(len(train_data) * 0.3)
```

```python
# Sample the dataset
sampled_data = train_data[:num_images_to_sample]
```

```python
len(sampled_data)
```

```python
# Define the output pickle file path
output_pickle_file = "/content/drive/MyDrive/Research Study/Data_RandomSampled_30%.pkl"

# Save the sampled data to a pickle file
with open(output_pickle_file, 'wb') as f:
    pickle.dump(sampled_data, f)
```
